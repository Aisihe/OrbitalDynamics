```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
#Simple Oscillator
k = 1
m = 1
x = np.linspace(-10, 10, 100)

for i in range(-30, 300, 10):
    v = -np.sqrt(k * x[0:50]**2 / m + i)
    v = np.concatenate((v, np.sqrt(k * x[50:100]**2 / m + i)))
    plt.plot(x, v, color='blue')
```

```python
# parameters
m = 1.0     # mass
k = 1.0     # spring constant
bound = 10
x = np.linspace(-bound, bound, 400)

# choose different total energies
energies = range(10)

```

```python
plt.figure(figsize=(7,7))
for E in energies:
    v = np.sqrt(np.clip((2*E)/m + (k/m)*x**2, 0, None))
    plt.plot(x, v, 'b')
    plt.plot(x, -v, 'b')


plt.axhline(0, color='k', linewidth=0.8)
plt.axvline(0, color='k', linewidth=0.8)

plt.title("Phase Diagram: Repulsive Spring")
plt.xlabel("x")
plt.ylabel("v")
plt.grid(True)
plt.show()
```
